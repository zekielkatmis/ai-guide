## GroupBy

```python
newDictionary = {"Departmanlar" : ["Yazlm", "Yazlm","Pazarlama","Pazarlama","HR", "HR"],
         "Elemanlar" : ["Ahmet","Mehmet","Zeki","Enes","Furkan","Eyan"],
         "Maas" : [100,150,200,250,300,400]}
maasDataFrame = pd.DataFrame(newDictionary)
maasDataFrame
```

## Concat

```python
dic1 = {"Isim": ["Ahmet","Mehmet", "Zeynep"],
        "Spor":["Kosu","Yuzme","Basket"],
       "Kalori" : [400,600,780]}
df1 = pd.DataFrame(dic1)
df1
```

```python
dic2 = {"Isim": ["Zeki","Elif", "Esra","Avni"],
        "Spor":["Kosu","Yuzme","Basket","Futbol"],
       "Kalori" : [400,600,780,800]}
df2 = pd.DataFrame(dic2)
df2
```

```python
dic3 = {"Isim": ["Enes","Furkan"],
        "Spor":["Kosu","Yuzme"],
       "Kalori" : [400,600]}
df3 = pd.DataFrame(dic3)
df3
```

```python
result=pd.concat([df1,df2,df3])#axis=1 verme durumunda yan yana birletirir
result
```

```python
result.reset_index(inplace=True)
```

```python
result.drop("index",axis=1,inplace=True)
result
```

## Pivot

```python
newData = {"Karakter Snf" : ["Cars", "Cars","Simpson","Simpson","Simpson"],
          "Karakter Ismi" : ["Cartman","Kenny","Homer","Bart","Bart"],
          "Karakter Yas" : [20,50,23,31,69]}
karakterDF=pd.DataFrame(newData)
karakterDF
```

```python
karakterDF.pivot_table(values="Karakter Yas", index=["Karakter Snf","Karakter Ismi"],aggfunc=np.sum)
# aggfunc ayn karakterlerin yaslarna uygulanacak ilemi belirler
# bir ey yazmazsak ya ortalamasn yazard
```
