```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
%matplotlib inline 
#plt.show() yazmamak i√ßin
```

```python
age = [10,25,30,42,55,67,79,85,93]
weight = [70,50,88,66,81,78,98,100,62]
```

```python
numpyAge = np.array(age)
numpyWeight = np.array(weight)
```

```python
plt.plot(numpyAge,numpyWeight,"g") #g rengin ne olacaƒüƒ±nƒ± belirtir
plt.xlabel("Age")
plt.ylabel("Weight")
plt.title("Age/Weight")
plt.show()
```

```python
numpyArray = np.linspace(0,10,20)
numpyArray
```

```python
numpyArray2 = numpyArray ** 3
numpyArray2
```

```python
plt.plot(numpyArray,numpyArray2,"g*-") 
```

```python
plt.subplot(1,2,1)
plt.plot(numpyArray,numpyArray2,"r*-")
plt.subplot(1,2,2)
plt.plot(numpyArray2,numpyArray,"g--")
```

```python
newFigure = plt.figure()
figureAxes = newFigure.add_axes([0.1,0.1,0.9,0.9])
# add_axes(-,-, plotWidth, plotHeight)
figureAxes.plot(numpyArray,numpyArray2,"g")
figureAxes.set_xlabel("Age")
figureAxes.set_ylabel("Weight")
figureAxes.set_title("Age/Weight")
```

```python
figure2 = plt.figure()

eksen1 = figure2.add_axes([0.1,0.1,0.9,0.9])
eksen2 = figure2.add_axes([0.2,0.6,0.3,0.3])

eksen1.plot(numpyArray,numpyArray2,"g")
eksen1.set_xlabel("x ekseni")
eksen1.set_ylabel("y ekseni")
eksen1.set_title("Ana Grafik")

eksen2.plot(numpyArray2,numpyArray,"b")
eksen2.set_xlabel("x ekseni")
eksen2.set_ylabel("y ekseni")
eksen2.set_title("Sub Grafik")
```

```python
(figure3, eksen3) = plt.subplots(nrows=1, ncols=2)
type(eksen3) #numpy.ndarray

for eksenler in eksen3:
    eksenler.plot(numpyArray,numpyArray2,"g")
    
plt.tight_layout() # plotlarƒ± birbirinden ayƒ±racaktƒ±r
```

```python
newFigure = plt.figure()

eksen4 = newFigure.add_axes([0.1,0.1,0.9,0.9])

eksen4.plot(numpyArray, numpyArray ** 2, label = "numpyArray ** 2")
eksen4.plot(numpyArray, numpyArray ** 3, label = "numpyArray ** 3")
eksen4.plot(numpyArray, numpyArray ** 3.1, label = "numpyArray ** 4")
eksen4.legend(loc=0) #loc legendin yerini deƒüi≈ütiriyor
```
